### Random walk

Hint
To create a list with a single zero, use [0].

Use range(100) to make the loop run for 100 times.

step = random_walk[-1] sets step to the last element in random_walk.

To print out a variable x, type print(x) on a new line.

If your code takes too long to execute, make sure you are iterating over the correct range.

```python
# NumPy is imported, seed is set

# Initialize random_walk
np.random.seed(123)
random_walk = [0]

# Complete the ___
for x in range(100) :
    # Set step: last element in random_walk
    step = random_walk[-1]

    # Roll the dice
    dice = np.random.randint(1,7)

    # Determine next step
    if dice <= 2:
      # Replace below: use max to make sure step can't go below 0
        # step = step - 1
        step = max(0, step - 1)
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    # append next_step to random_walk
    random_walk.append(step)

# Print random_walk
print(random_walk)
```

```python
# Initialize all_walks (don't change this line)
all_walks = []

# Simulate random walk five times
for i in range(5) :

    # Code from before
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)

        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)

    # Append random_walk to all_walks
    all_walks.append(random_walk)

# Print all_walks
print(all_walks)


'''
\
This solution solves the exercise by following the instructions step by step.

First, it initializes an empty list all_walks. This list will be used to store all the random walks.

Next, it simulates a random walk five times using a for loop. The range(5) function generates a sequence of numbers from 0 to 4, which is used to repeat the loop five times.

Inside this loop, it initializes a list random_walk with a single element 0. This list will store the steps of a single random walk.

Then, it simulates a random walk of 100 steps using another for loop. The range(100) function generates a sequence of numbers from 0 to 99, which is used to repeat the loop 100 times.

In each iteration of this loop, it first gets the last step from random_walk and stores it in step. Then, it generates a random number between 1 and 6 (inclusive) using np.random.randint(1,7) and stores it in dice.

Depending on the value of dice, it updates step accordingly. If dice is less than or equal to 2, it decreases step by 1 but not below 0. If dice is between 3 and 5 (inclusive), it increases step by 1. If dice is 6, it increases step by a random number between 1 and 6 (inclusive).

After updating step, it appends step to random_walk.

After simulating a random walk of 100 steps, it appends random_walk to all_walks.

Finally, after simulating a random walk five times, it prints all_walks. This will display a list of five lists, each containing 101 elements (the initial 0 plus 100 steps).

Thank you for your feedback!
'''
```

```python
# numpy and matplotlib imported, seed set.

# initialize and populate all_walks
all_walks = []
for i in range(5) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)
    all_walks.append(random_walk)

# Convert all_walks to NumPy array: np_aw
np_aw = np.array(all_walks)

# Plot np_aw and show
plt.plot(np_aw)
plt.show()

# Clear the figure
plt.clf()

# Transpose np_aw: np_aw_t
np_aw_t = np.transpose(np_aw)

# Plot np_aw_t and show
plt.plot(np_aw_t)
plt.show()
```
